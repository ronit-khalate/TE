Data Visualization II

Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names : 'sex' and 'age')


Import required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load Dataset

```python
df= pd.read_csv("Titanic-Dataset.csv")
df
```

Preprocessing data set

```python
required_df = df[["Name","Sex","Age","Survived"]]
required_df
```

```python
required_df.info()
```

```python
required_df.isnull().sum()
```

```python
proc_df = required_df.dropna()
proc_df
```

```python
proc_df.isnull().sum()
```

Grouping data and ploting box plot

```python
df2 = proc_df.groupby(["Sex","Survived"])
df2

for m , f in df2:
    print(m)
    print(f)
```

```python
p=df2.get_group(("male",1))
p
```

```python
plt.title("Representation of age wr male who survived") 
plt.boxplot(p["Age"])
print(p["Age"].mean())
```

```python
q= df2.get_group(("male",0))
q
```

```python
plt.title("Representation of age wr male who not survived") 
plt.boxplot(q["Age"])
print(q["Age"].mean())
```

```python
r= df2.get_group(("female",1))
r
```

```python
plt.title("Representation of age wr female who survived") 
plt.boxplot(r["Age"])
print(r["Age"].mean())
```

```python
s=df2.get_group(("female",0))
s
```

```python
plt.title("Representation of age wr female who not survived") 
plt.boxplot(s["Age"])
print(s["Age"].mean())
```
